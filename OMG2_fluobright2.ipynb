```python
import skimage.io
import skimage.color
import skimage.transform
import numpy as np
import matplotlib.pyplot as plt
```

```python
worm1b = skimage.io.imread(fname="/Users/ykang/ImgBasics/OMG2_fluobright2/OMG2_fluobright2_t001.tif")
fig, ax = plt.subplots()
plt.imshow(worm1b)
```

```python
worm1b[worm1b < 30] = 0
worm1b[worm1b >= 30] = 255
worm1b[worm1b==255]=128  
worm1b[worm1b==0]=255 
worm1b[worm1b==128]=0
fig,ax=plt.subplots()
plt.imshow(worm1b)
```

```python
worm1h = skimage.io.imread(fname="/Users/ykang/ImgBasics/OMG2_fluobright2/OMG2_fluobright2_t001.tif")
worm1h[worm1h >180] = 255
worm1h[worm1h <= 180] = 0
fig,ax=plt.subplots()
plt.imshow(worm1h)
skimage.io.imsave(fname="/Users/ykang/ImgBasics/OMG2_fluobright2/head/OMG2_fluobright2_t001_head.tiff", arr=worm1h)
```

```python
worm1 = worm1h+worm1b
fig,ax=plt.subplots()
plt.imshow(worm1)
skimage.io.imsave(fname="/Users/ykang/ImgBasics//OMG2_fluobright2/body/OMG2_fluobright2_t001_body.tiff", arr=worm1)
```
